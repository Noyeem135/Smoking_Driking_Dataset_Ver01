```python
df["sex"].replace({"Male":1, "Female":0},inplace=True)
```

```python
df["DRK_YN"].replace({"Y":1, "N":0},inplace=True)
```

```python
# Dividing dependent and independent variables
X = df.drop("DRK_YN", axis=1)
y = df["DRK_YN"]
```

```python
### Sampling
# from imblearn.over_sampling import SMOTE 
# from collections import Counter

# sampler = SMOTE()
# X, y = sampler.fit_resample(X, y)

# print('SMOTE {}'.format(Counter(y)))
```

```python
#Train_Test_Splitting
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
dt_model = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = dt_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, plot_roc_curve
```

```python
print(classification_report(y_test, y_pred))
```

```python
dt_model.score(X_test,y_test)
```

```python
#â–ª Accuracy = (TP + TN ) / (TP + FP + TN + FN) 
accuracy_score(y_test, dt_model.predict(X_test))
```

```python
cm = confusion_matrix(y_test, dt_model.predict(X_test))
cm
```

```python
(64170+63891)/(64170+35425+34784+63891)
```

```python
sns.heatmap(cm, annot=True)
```

```python
# Precision or Positive Predictive Value (PPV) 
# Precision = TP / ( TP + FP) 
precision_score(y_test, dt_model.predict(X_test))
```

```python
#show formula
recall_score(y_test, dt_model.predict(X_test))
```

```python
# False Positive Rate (FPR) or 1-Specificity 
# FPR = False Positive / (False Positive + True Negative)
# FPR = FP/ (FP/TN)

f1_score(y_test, dt_model.predict(X_test))
```

```python
plot_roc_curve(dt_model, X_test, y_test)
plt.plot([0,1],[0,1])
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier

clf= KNeighborsClassifier(n_neighbors=25)

# Train the model using the training sets
knn_model=clf.fit(X_train,y_train)

y_pred=knn_model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```
